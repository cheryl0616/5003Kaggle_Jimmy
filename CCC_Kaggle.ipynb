```python
import pandas as pd
import numpy as np
```

```python
#get data(please replace the data source path with your own data path)
data_train = pd.read_csv("E:/Cheryl/Class/5001 foundatains of data analytics/Kaggle_Jimmygame/data/train.csv")
data_train = data_train.set_index("id")
data_test = pd.read_csv("E:/Cheryl/Class/5001 foundatains of data analytics/Kaggle_Jimmygame/data/test.csv")
data_test = data_test.set_index("id")
data_test[data_test.isnull().purchase_date==True]
```

```python
# Handling the missing data
data_train=data_train.dropna(how='any', axis = 0)
data_test = data_test.fillna(value={"purchase_date": data_test["purchase_date"].mode()[0], 
                                    "total_positive_reviews": data_test["total_positive_reviews"].mode()[0], 
                                    "total_negative_reviews":data_test["total_negative_reviews"].mode()[0]})
```

```python
all_data = data_train.append(data_test,sort=False)
all_data
```

```python

# Handling genres,categories,tags
dummies_genres= all_data["genres"].str.get_dummies(",")
dummies_categories= all_data["categories"].str.get_dummies(",")

df = pd.concat([all_data, dummies_genres, dummies_categories], axis=1)
df = df.drop(["genres","categories","tags"],axis=1)
df
```

```python
# Handling time
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
df['purchase_dayofweek'] = pd.to_datetime(df['purchase_date']).dt.dayofweek
df['purchase_year'] = pd.to_datetime(df['purchase_date']).dt.year
df['purchase_month'] = pd.to_datetime(df['purchase_date']).dt.month
df['release_date'] = pd.to_datetime(df['release_date'])
df['release_dayofweek'] = pd.to_datetime(df['release_date']).dt.dayofweek
df['release_year'] = pd.to_datetime(df['release_date']).dt.year
df['release_month'] = pd.to_datetime(df['release_date']).dt.month
df['release_purchase_interval'] = df["purchase_date"]-df["release_date"]
df['release_purchase_interval'] = df['release_purchase_interval'].dt.days

df = df.drop(["purchase_date","release_date"],axis=1)
df
```

```python
# Handling review info

df["total_positive_reviews_ratio"]=df["total_positive_reviews"]/(df["total_positive_reviews"]+df["total_negative_reviews"])
df['total_positive_reviews_ratio'] = df['total_positive_reviews_ratio'].map(lambda x: x*100)  
df["total_negative_reviews_ratio"]=df["total_negative_reviews"]/(df["total_positive_reviews"]+df["total_negative_reviews"])
df['total_negative_reviews_ratio'] = df['total_negative_reviews_ratio'].map(lambda x: x*100)  
df = df.fillna(0)
df = df.drop(["total_positive_reviews","total_negative_reviews"],axis=1)
df
```

```python

df2 = df.tail(90)
df1 = df.head(355)
train_data_y = df1.iloc[:,0:1]
df = df.drop(["playtime_forever"],axis=1)
df
```

```python

# nomalization
from sklearn.preprocessing import StandardScaler
s_df1 = StandardScaler()
df_nor = s_df1.fit_transform(df)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold,cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import linear_model

# train model
RF_model = RandomForestRegressor(n_estimators=100,max_depth=2)
training_data = df_nor[:355, :]
test_data = df_nor[-90:, :]
training_x ,testing_x,training_y,testing_y = train_test_split(training_data,train_data_y,test_size=0.2,random_state=0)
RF_model.fit(training_x, training_y)

'''
GB_model = GradientBoostingRegressor()
x = df_train[:, 1:]
y = df_train[:, 0]
GB_model.fit(x, y)

LE_model =  LinearRegression()
x = df_train[:, 1:]
y = df_train[:, 0]
LE_model.fit(x, y)


ET_model = ExtraTreesRegressor()
x = df_train[:, 1:]
y = df_train[:, 0]
ET_model.fit(x, y)
'''



RF_predict_y = RF_model.predict(testing_x)
ran_rmse = mean_squared_error(testing_y, RF_predict_y)
ran_rmse = np.sqrt(ran_rmse)
print(ran_rmse)
```

```python
# prediction
predictions = RF_model.predict(test_data)
predictions
```
